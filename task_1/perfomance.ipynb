```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
task = np.array([100, 250, 500, 1000, 1500, 2000])

time_obvious = np.array([5375, 102729, 2587215, 21594821, 0, 0])
time_1thr = np.array([1178384, 1423438,1847944 , 10810710 , 42740508, 82617775])
time_2thr = np.array([1192621, 1305297, 1431671, 4888188, 18695460, 34001851])
time_4thr = np.array([1176340, 1388251, 2200345, 7450264, 14064767, 31294135])
time_8thr = np.array([])


plt.figure(dpi=200)
plt.plot(task, time_obvious, '-bo', label="obvious")
plt.plot(task, time_1thr, '-ro', label="1 Thread")
plt.plot(task, time_2thr, '-go', label="2 Thread")
plt.plot(task, time_4thr, '-yo', label="4 Thread")
plt.xlabel('task ( matrix x*x )')
plt.ylabel('clean time in microSec ')
plt.title('SpeedUp')
plt.legend()
plt.show()
plt.savefig("SpeedUp_plot")
```
